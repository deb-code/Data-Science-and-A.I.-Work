```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

## Univariate Analysis

```python
df_setosa = df.loc[df['species']=='setosa']
```

```python
df_verginica = df.loc[df['species']=='verginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_verginica['sepal_length'],np.zeros_like(df_verginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.ylabel('Univariate')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue = "species", size = 5).map(plt.scatter,"petal_length","petal_width").add_legend()
```

## Mulltivariate Analysis

```python
sns.pairplot(df,hue = "species",size = 5)
```
